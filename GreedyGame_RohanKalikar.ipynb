Import the libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

Read the given data into Python

```python
data = pd.read_csv("/content/sample_data_set_2 - Sheet1 (1).csv")
data = data.dropna()
data = data.reset_index(drop=True)
print(data)
```

```python
# Drop some columns
data_new = data.drop(['Fill Rate','Render Rate','State'], axis = 1)

# Group the dataset
data2 = data_new.groupby(['Date','App','Ad Requests'],as_index=False).sum(['Revenue','Ad Responses','Impressions','1st Clicks','2nd Clicks'])
```

```python
# Now group by 'App'
data2 = data2.groupby(['App'],as_index=False).sum()
```

```python
data2.head()
```

```python
#Enter the fill rates and Render rates
data2['Fill Rate'] = (data2['Ad Responses']/data2['Ad Requests'])*100
data2['Render Rate']= (data2['Impressions']/data2['Ad Responses'])*100

data2 = data2.dropna()
```

Correlation Matrix without Ad Groups

```python
data2.corr()
```

K-Means Clustering Classification- without Ad Groups


```python
x = data2[['Revenue']]


#Find Number of clusters
wcss=[]
for i in range(1,7):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)

number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow title')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
#2


kmeans = KMeans(2)
kmeans.fit(x)


identified_clusters = kmeans.fit_predict(x)
identified_clusters



data_with_clusters = data2.copy()
data_with_clusters['Clusters'] = identified_clusters

```

```python
#Sort the new data with clusters
data_with_clusters.sort_values(by='Clusters', ascending = False)
```

Dataset with Ad Groups

```python
data_groups = data_new.groupby(['Ad Group'],as_index=False).sum()
data_groups['Fill Rate'] = (data_groups['Ad Responses']/data_groups['Ad Requests'])*100
data_groups['Render Rate']= (data_groups['Impressions']/data_groups['Ad Responses'])*100
data_groups = data_groups.dropna()
```

K-Means Clusterinig Classification with Ad Groups

```python
y = data_groups[['Render Rate']]

kmeans_g = KMeans(2)
kmeans_g.fit(y)

identified_clusters_g = kmeans_g.fit_predict(y)
identified_clusters_g

data_with_clusters_g = data_groups.copy()
data_with_clusters_g['Clusters'] = identified_clusters_g

data_with_clusters_g.sort_values(by='Clusters', ascending = True)
```
